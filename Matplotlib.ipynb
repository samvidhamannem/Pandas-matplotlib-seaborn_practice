```python
x=np.arange(1,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,'b*-')
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Num vs Squares',loc='center')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g*-')
plt.subplot(1,2,2)
plt.plot(y,x,'y*-')
plt.show()
```

```python
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)

plt.subplot(2,2,1)
plt.plot(x,y1,'r*-')
plt.xlabel('x')
plt.ylabel('y1')
plt.title('$y_1=x$')

plt.subplot(2,2,2)
plt.plot(x,y2,'g*-')
plt.xlabel('x')
plt.ylabel('y2')
plt.title('$y_2=x^2$')

plt.subplot(2,2,3)
plt.plot(x,y3,'b*-')
plt.xlabel('x')
plt.ylabel('y3')
plt.title('$y_3=x^3$')

plt.subplot(2,2,4)
plt.plot(x,y4,'g*-')
plt.xlabel('x')
plt.ylabel('y4')
plt.title('$y_4=/sqrt{x}$')

plt.show()

```
