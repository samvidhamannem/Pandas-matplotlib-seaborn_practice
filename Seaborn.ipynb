```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
sns.jointplot(x='total_bill',y='size',data=tips,kind='hex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.barplot(x='smoker',y='tip',data=tips,hue='sex',estimator=sum)
```

```python
sns.countplot(x='sex',data=tips)
```
